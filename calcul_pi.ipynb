### Càcul de pi
Si tenim un polígon circumscrit en una circumferència de radi 1. <br/><br/>
Si fem tendir a $\infty$ el nombre de costats &rarr; el poligon es converteix en una circumferència.
- $A_{\text{circumferència}} = \pi\,r^2$
- $A_{\text{poligon}} = n·A_{\text{triangle}}$
- $A_{\text{triangle}} = \frac{b·h}{2}$
- on: 
    - n = número de costats del polígon
    - r = radi &rarr; r = 1

- Area triangle: 
    - $\alpha = \frac{360}{n}$
    - $\frac{b}{2} = 1·sin(\frac{360}{2n}) \rightarrow b = 2·sin(\frac{360}{2n})$
    - $1^2 = h^2 + sin^2(\frac{360}{2n}) \rightarrow h = cos(\frac{360}{2n})$ 
        - $1 = cos^2(x) + sin^2(x)$


<br/>

- $A_h = n · A_t = n · \frac{b·h}{2} = n · \frac{2sin(\frac{360}{2n})cos(\frac{360}{2n})}{2}$
- $A_h(n) = n·sin(\frac{360}{2n})cos(\frac{360}{2n})$
    - on $n$ és el nombre de costats del poligon.

<br/>

- Llavors:
- Limit quan: $n \rightarrow \infty$ per $A_h(n)$
$$
\lim_{n \to \infty} A_h(n)
$$
<br/>
$$
\lim_{n \to \infty} n·sin(\frac{360}{2n})cos(\frac{360}{2n})
$$
<br/>
$$
= \infty·sin(\frac{360}{\infty})cos(\frac{360}{\infty})
$$
<br/>
$$
= \infty·sin(0)cos(0)
$$
<br/>
$$
= \infty·0·1
$$
<br/>
$$
= 0
$$

#### Desepció!
- Hauría d'haber sigut: 
$$
\lim_{n \to \infty} A_h(n) = r^2 \pi
$$
- r = 1
$$
\lim_{n \to \infty} A_h(n) = \pi
$$

### Provem amb el perimetre: 
- Mateix concepte però la formula del perimetre.
- $P_{\text{circumferència}} = 2\pi\,r$
- $P_{\text{poligon}} = n·b_{\text{triangle}}$
    - b = base del triangle
    - n = nombre de costats del poligon circumscrit
- Llavors, si busquem la formula per calcular el perimetre del poligon i li apliquem el limit de n a inf. 
    - Hauríem de poder igualar la formula del perimetre de la circumferència a aquest limit. 

- $A_p(n) = n·b$
    - on $n$ és el nombre de costats del poligon.
    - $b = 2·sin(\frac{360}{2n})$
    
- $\lim_{n \to \infty} n·2·sin(\frac{360}{2n}) = \infty · 0 = 0$


### Podem probar de fer el límit amb python
- En comptes de fer tendir la formula a infinit: 
$$
\lim_{n \to \infty} A_h(n) = r^2 \pi
$$
- r = 1
- Calcularem el valor de 
- $A_h(n) =\pi$ 
- per n molt grans i veurem si es compleix la igualtat. 

$$A_h(n) = n·sin(\frac{360}{2n})cos(\frac{360}{2n})$$

```python
import math
import numpy as np

def calcular_area(n):
    x = math.radians(360)/(2*n)
    return n *math.sin(x) *math.cos(x)

calcular_area(6)
```

```python
n_poligons = 1000

areas = [calcular_area(n) for n in range(3, n_poligons +1)]

areas[-5:]
```

```python
# a la llibreria math, tenim pi amb 15 decimals
math.pi, len(str(math.pi))
```

```python
# mirem l'error que estem fent:
np.array(areas[-5:]) -math.pi
```
