```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir el rango de valores para x
x = np.linspace(-1, 1, 10)
```

```python
def plot(x, y):
    # Crear la gráfica
    plt.figure(figsize=(4, 3))
    plt.plot(x, y, '-b')

    # Etiquetas de los ejes
    plt.xlabel('x')
    plt.ylabel('y')

    # Mostrar la gráfica
    plt.axhline(0, color='black',linewidth=0.3)
    plt.axvline(0, color='black',linewidth=0.3)
    plt.show()
```

#### Función
- $y = 2x + 2$

```python
# Calcular los valores correspondientes de y
y = 2*x + 1
plot(x, y)
```

- $y=x$

```python
y = x
plot(x, y)
```

```python
# Definir el rango de valores para x
x = np.linspace(-1, 1, 10)

# Calcular los valores correspondientes de y para ambas funciones
y1 = 2*x + 1
y2 = x + 1

# Crear la gráfica
plt.figure(figsize=(8, 6))

# Graficar y = 2x + 2
plt.plot(x, y1, '-r', label='y = 2x + 1')

# Graficar y = x + 10
plt.plot(x, y2, '-b', label='y = x + 1')

# Etiquetas de los ejes
plt.xlabel('x')
plt.ylabel('y')

# Título de la gráfica
plt.title('Gráfica de y = 2x + 1 y y = x + 1')

# Agregar la leyenda
plt.legend(loc='upper left')

# Mostrar la gráfica
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-1, 1, 10)

# Calcular los valores correspondientes de y para ambas funciones
y1 = 2*x
y2 = x

# Crear la gráfica
plt.figure(figsize=(8, 6))

# Graficar y = 2x
plt.plot(x, y1, '-r', label='y = 2x')

# Graficar y = x
plt.plot(x, y2, '-b', label='y = x')

# Etiquetas de los ejes
plt.xlabel('x')
plt.ylabel('y')

# Título de la gráfica
plt.title('Gráfica de y = 2x y y = x')

# Especificar los límites de los ejes x e y
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Agregar la leyenda
plt.legend(loc='upper left')

# Mostrar la gráfica
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los coeficientes de la función cúbica
a = 1
b = 2
c = -3
d = -2  # Ajustar d para mover el punto de corte en y

# Definir el rango de valores para x
x = np.linspace(-10, 10, 400)

# Calcular los valores correspondientes de y para la función cúbica
y = a*x**3 + b*x**2 + c*x + d

# Crear la gráfica
plt.figure(figsize=(8, 6))
plt.plot(x, y, '-r', label=f'y = {a}x^3 + {b}x^2 + {c}x + {d}')

# Etiquetas de los ejes
plt.xlabel('x')
plt.ylabel('y')

# Título de la gráfica
plt.title('Gráfica de y = ax^3 + bx^2 + cx + d con punto de corte en y = -2')

# Agregar la leyenda
plt.legend(loc='upper left')

# Mostrar la gráfica
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Definir los valores de x
x = np.linspace(-5, 5, 100)

# Calcular los valores de y
y = x**2

# Crear un DataFrame de pandas
data = pd.DataFrame({'x': x, 'y': y})

# Configurar el estilo de Seaborn
sns.set(style="whitegrid")

# Crear el gráfico utilizando Seaborn
plt.figure(figsize=(8, 6))
sns.lineplot(data=data, x='x', y='y')

# Etiquetas de los ejes
plt.xlabel('x')
plt.ylabel('y')

# Título de la gráfica
plt.title('Gráfica de y = x^2')

# Mostrar la gráfica
plt.show()
```
