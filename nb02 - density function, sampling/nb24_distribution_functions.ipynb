### cumulative density functions (ecdf, cdf)

- An empirical cumulative distribution function (ecdf) represents the proportion or count of observations falling below each unique value in a dataset.
- A cumulative distribution function (cdf) represents the cummulated probability density below a given value.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_theme() # Setting seaborn style
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.columns
```

#### ecdf/cdf 

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
# empirical cumulative distribution function
sns.ecdfplot(ax = axs[0], data = iris, x = "petal_length")
# cumulative distribution function
sns.kdeplot(ax = axs[1], data = iris, x = "petal_length", bw_adjust = 0.2, cumulative = True)
# both
sns.ecdfplot(ax = axs[2], data = iris, x = "petal_length")
sns.kdeplot(ax = axs[2], data = iris, x = "petal_length", bw_adjust = 0.2, cumulative = True);
```

La segona és amb el kernel

#### cdf computation using sklearn

- same as with the pdf:
    - seaborn helps us to plot the ecdf/cdf but does not return it
    - if we need the cdf we can compute it by integrating the pdf

```python
from sklearn.neighbors import KernelDensity
```

```python
# first step is computing the pdf
data, bandwidth = iris.sepal_length, 0.2
kde = KernelDensity(kernel = 'gaussian', bandwidth = bandwidth).fit(np.array(data)[:, None])
# crear les mostres o els nous valors que volem tenir/inferir, farem pdf(x) i ens donarà resultat
bin_width = 0.5
bins = np.round(np.arange(iris['petal_length'].min() -0.4, iris['petal_length'].max() +0.4, step = bin_width), 2)
# densities
pdf = np.exp(kde.score_samples(bins[:, None]))
```

```python
# second step is to compute the cummulative values using the np.cumsum() function
cdf = np.cumsum(pdf *bin_width)
```

- Està multiplicant per la longitud del bin perque sumi 1. perque retorna densitat i volem probabilitat.
- Ja que per fer el número random després, hem d'estar dins l'interval de 1. Perquè volem la probabilitat de trobar un valor, no la densitat que tenim d'ell. 
- `cdf` densitats acomulades per cada punt.

```python
import pandas as pd
pd.DataFrame(np.stack((bins, pdf, cdf)).T, columns = ['bin', 'pdf', 'cdf'])
```

#### define our own cdf function

```python
def mycdf(data, kernel = 'gaussian', bandwidth = 0.2, bin_width = 0.2):
    kde = KernelDensity(kernel = 'gaussian', bandwidth = bandwidth).fit(np.array(data)[:, None])
    bins = np.round(np.arange(data.min() -0.4, data.max() +0.4, step = bin_width), 2)
    pdf = np.exp(kde.score_samples(bins[:, None]))
    cdf = np.cumsum(pdf *bin_width)
    return bins, pdf, cdf
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# compute bins, pdf, cdf
kernel, bandwidth, bin_width = 'gaussian', 0.2, 0.2
bins, pdf, cdf = mycdf(iris.petal_length, kernel = kernel, bandwidth = bandwidth, bin_width = bin_width)

# seaborn density histogram
sns.histplot(ax = axs[0], data = iris, x = "petal_length", binwidth = 0.2, stat = 'density')
# seaborn pdf
sns.kdeplot(ax = axs[0], data = iris,  x = 'petal_length', bw_adjust = 0.2)
# sklearn pdf
sns.lineplot(ax = axs[0], x = bins, y = pdf)

# seaborn ecdf
sns.ecdfplot(ax = axs[1], data = iris, x = "petal_length")
# seaborn cdf
sns.kdeplot(ax = axs[1], data = iris, x = "petal_length", bw_adjust = 0.2, cumulative = True)
# sklearn cdf
sns.lineplot(ax = axs[1], x = bins +0.1, y = cdf);
```

- És la probabilitat de que el valor sigui x o menor. si ho fem al revés 0.8 -> dona 5.5, el 80% de valors seran 5.5 o menys.
- Per tant si fem numeros aleatoris la proporcio amb la que samplejem serà correcte.(% de observar el valor).
- Per això ho hem convertit a probabilitat/densitat (el que he escrit a dalt).
- Això és inverse sampling.
- Si agafem un punt on és pla perquè sigui 0, donarem el primer (el de més a la esquerre). Per tant no sortiràn valors que no hi havia. 
- Si volem aproximar la taronja a la blava, les dades han de ser representatives de la realitat. Que cap area del domini no surti al dataset que tenim. 
- Ho estem fent amb les variables per separat i en realitat és fa per totes (multivariant).
- Per norma general està bé ser conservador o esceptic amb les dades que tenim.

### Posem les dades com a parametre per poder fer proves:

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# compute bins, pdf, cdf
kernel, bandwidth, bin_width = 'gaussian', 0.1, 0.2
col = 'sepal_length'
bins, pdf, cdf = mycdf(iris[col], kernel = kernel, bandwidth = bandwidth, bin_width = bin_width)

# seaborn density histogram
sns.histplot(ax = axs[0], data = iris, x = col, binwidth = 0.2, stat = 'density')
# seaborn pdf
sns.kdeplot(ax = axs[0], data = iris,  x = col, bw_adjust = 0.2)
# sklearn pdf
sns.lineplot(ax = axs[0], x = bins, y = pdf)

# seaborn ecdf
sns.ecdfplot(ax = axs[1], data = iris, x = col)
# seaborn cdf
sns.kdeplot(ax = axs[1], data = iris, x = col, bw_adjust = 0.2, cumulative = True)
# sklearn cdf
sns.lineplot(ax = axs[1], x = bins +0.1, y = cdf);
```
