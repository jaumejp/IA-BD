### descriptive statistics with seaborn

- data distribution:
    - boxplot
    - histogram
- correlation:
    - pair plot

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme() # Setting seaborn style
```

```python
iris = sns.load_dataset("iris")
```

#### class conditional feature box-plot

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Iris class conditional boxplots')
sns.boxplot(ax = axes[0, 0], data = iris, x= 'species', y ='sepal_length')
sns.boxplot(ax = axes[0, 1], data = iris, x= 'species', y ='sepal_width')
sns.boxplot(ax = axes[1, 0], data = iris, x= 'species', y ='petal_length')
sns.boxplot(ax = axes[1, 1], data = iris, x= 'species', y ='petal_width');
```

#### histogram

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Iris class conditional histograms')
sns.histplot(ax = axes[0, 0], data = iris, x = "sepal_length", hue = "species", binwidth = 0.2)
sns.histplot(ax = axes[0, 1], data = iris, x = "sepal_width", hue = "species", binwidth = 0.2)
sns.histplot(ax = axes[1, 0], data = iris, x = "petal_length", hue = "species", binwidth = 0.2)
sns.histplot(ax = axes[1, 1], data = iris, x = "petal_width", hue = "species", binwidth = 0.2);
```

#### pair plot 

- grid figure with:
    - scatterplots (pairwise joint distribution, correlation)
    - histograms in the diagonal (self distribution)

```python
sns.pairplot(iris, hue = "species", diag_kind = "hist");
```
