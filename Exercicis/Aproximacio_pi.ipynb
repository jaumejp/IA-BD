```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total_shots = 150
x = np.random.rand(total_shots)
y = np.random.rand(total_shots)
```

```python
figure, axes = plt.subplots()

plt.scatter(x, y, label='Punts')

draw_circle = plt.Circle((0, 0), 1, fill=False, edgecolor='red', label='Cercle')
axes.add_artist(draw_circle)

axes.set_aspect(1)

plt.xlabel('Eix X')
plt.ylabel('Eix Y')
plt.title('Representaci√≥ de la tirada de dards')

plt.legend()

plt.show()
```

```python
dardo_shot = np.sqrt(x**2 + y**2)
```

```python
dardo_in = dardo_shot <= 1
```

```python
pi = sum(dardo_in) / total_shots
pi * 4
```

```python
results = {}

for i in range(len(dardo_in)):
    results[i+1] = 4 * sum(dardo_in[:i+1]) / (i+1)
```

```python
tirades = results.keys()
resultat = results.values()

plt.plot(tirades, resultat)

plt.xlabel('Total Acomulat')
plt.ylabel('Pi')

plt.title('Pi en funci√≥ de les tirades')

plt.show()
```

```python
def pi_plot(total_shots = 150):
    dardo_in = np.sqrt(np.square(np.random.rand(total_shots)) + np.square(np.random.rand(total_shots))) <= 1
    
    results = {i + 1: 4 * sum(dardo_in[:i + 1]) / (i + 1) for i in range(len(dardo_in))}
    
    print(f"Pi amb {total_shots} iteracions: {sum(dardo_in) / total_shots * 4}")
    
    plt.plot(results.keys(), results.values())
    plt.xlabel('Total Acomulat')
    plt.ylabel('Pi')
    plt.title('Pi en funci√≥ de les tirades')
    plt.show()
```

```python
for i in [10, 100, 1_000, 1_500, 2_000, 10_000, 20_000, 50_000]:
    pi_plot(i)
```
