### Class definition

```python
class Car():

    def __init__(self, make, model, color, year, price):

        # class attributes

        self.make = make
        self.model = model
        self.color = color
        self.year = year
        self.price = price
```

##### class instantiation

```python
mycar = Car('KIA', 'Carnival', 'Beige', 2007, 4860)
mycar
```

##### check attributes

```python
mycar.make
```

```python
mycar.price
```

### Class methods

```python
class Car():

    def __init__(self, make, model, color, year, price):

        # class attributes

        self.make = make
        self.model = model
        self.color = color
        self.year = year
        self.price = price

        self.nofee = True

    # class methods

    def years(self):
        return 'this car is %3d years old' %(datetime.now().year -self.year)

    def finance(self, months = 12, interest_rate = 5):
        if self.nofee:
            return 'This car is not financiable'
        else:
            return 'months: %3d, montly payment: %8.2f€' %(months, self.price *(1 + interest_rate /100) /months)
```

##### define instance of Car()

```python
yourcar = Car('BMW', 'Serie 5', 'Verde', 2001, 1800)
yourcar
```

```python
yourcar.years()
```

##### check methods

```python
yourcar.finance()
```

```python
yourcar.nofee = False
yourcar.finance()
```

```python
yourcar.finance(months = 18, interest_rate = 7.5)
```

### Class inheritance

```python
class BestCar(Car):

    def spot_price(self, percentage = 2.5):
        return 'Price %8.2f€, Spot price (-%4.2f%%) %8.2f€' %(self.price, percentage, self.price *(1 - percentage /100))
```

```python
hiscar = BestCar('BMW', 'Serie 5', 'Verde', 2001, 1800)
hiscar
```

##### BestCar() inherits all methods/attributes defined for its parent class Car()

```python
hiscar.years()
```

```python
hiscar.finance()
```

##### check BestCar() specific new methods

```python
hiscar.spot_price()
```

### Special methods

```python
class SuperCar(BestCar):

    # __repr__ method (instance representation)
    def __repr__(self):
        txt  = 'Company: %s \n' %self.make
        txt += '  model: %s \n' %self.model
        txt += '  color: %s \n' %self.color
        return txt

    # __str__ method (result of print(instance))
    def __str__(self):
        txt  = 'Company: %s \n' %self.make
        txt += '  model: %s \n' %self.model
        txt += '  color: %s \n' %self.color
        txt += '  price: %8.2f€ \n' % self.price
        return txt
```

- ***\_\_repr\_\_()***, defines how the class is shown

```python
mysupercar = SuperCar('Fiat', '500', 'Negro', 2017, 6490)
mysupercar
```

- ***\_\_str\_\_()***, defines how the clas is printed

```python
print(mysupercar)
```

##### SuperCar() inherits all methods/attributes defined for its parents classes BestCar() and Car()

```python
mysupercar, mysupercar.spot_price(3.0), mysupercar.finance(months = 18, interest_rate = 6.5)
```

```python
class Paralelogram:
  def __init__(self, altura, amplada):
    self.altura = altura
    self.amplada = amplada

  def area(self):
    return self.altura * self.amplada

  def perimetre(self):
    return (2 * self.amplada) + (2 * self.altura)

  def arestes(self):
    return 4

  def vertex(self):
    return 4
```

```python
class Paralelepipede(Paralelogram):
    def __init__(self, altura, amplada, gruix):
      super().__init__(altura, amplada)
      self.gruix = gruix

    def area(self):
      return self.altura * self.amplada * self.gruix

    def perimetre(self):
      return (2 * self.altura) + (2 * self.amplada) + (2 * self.gruix)

    def arestes(self):
      return 12

    def vertex(self):
      return 8
```

```python
class Cuadrat(Paralelogram):
    def __init__(self, costat):
      super().__init__(costat, costat)
```

```python
class Cub(Paralelepipede):
    def __init__(self, costat):
      super().__init__(costat, costat, costat)
```

```python
cuadrat = Paralelogram(1, 1)
print(cuadrat.area())
cuadrat.perimetre()
```
