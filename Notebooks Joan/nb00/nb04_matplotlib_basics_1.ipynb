### Matplotlib basics 1
- basic plotting functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### plotting a function $f(x)$

```python
x = np.linspace(-2, 2, 21)
x
```

```python
# f(x) = x
y1 = x
y1
```

```python
plt.plot(x, y1);
```

```python
# add vertical/horizontal axes lines
plt.plot(x, y1)
plt.axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
plt.axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5);
```

```python
# f(x) = |x|; absolute value
y0 = np.abs(x)
y0
```

```python
plt.plot(x, y1, 'b-')
plt.plot(x, y0, 'r-+', alpha = 0.5)
plt.axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
plt.axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5);
```

```python
# f(x) = x**2; square function
y2 = x**2
y2
```

```python
plt.plot(x, y1, 'b-')
plt.plot(x, y0, 'r-+', alpha = 0.6)
plt.plot(x, y2,'g-')
plt.axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
plt.axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5);
```

```python
# f(x) = x**3; cubic function
y3 = x**3
y3
```

```python
plt.plot(x, y1, 'b-')
plt.plot(x, y0, 'r-+', alpha = 0.6)
plt.plot(x, y2,'g-')
plt.plot(x, y3,'m-')
plt.axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
plt.axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5);
```

#### figures, multiplot

```python
fig, axs = plt.subplots(2, 2, figsize = (8, 8))
# fig is the canvas
fig.suptitle('Ttol de la figura')
# axs is a matrix, where each element is an axes object
axs[0, 1].set_title('subplot fila 0, col 1')
axs[1, 0].set_title('subplot fila 1, col 0')
axs;
```

```python
fig, axs = plt.subplots(2, 2, figsize = (8, 8))
fig.suptitle('functions of x')

axs[0, 0].set_title('f(x) = x')
axs[0, 0].plot(x, y1)
axs[0, 0].axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
axs[0, 0].axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5)

axs[0, 1].set_title('f(x) = |x|')
axs[0, 1].plot(x, y0)
axs[0, 1].axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
axs[0, 1].axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5)

axs[1, 0].set_title('f(x) = x**2')
axs[1, 0].plot(x, y2)
axs[1, 0].axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
axs[1, 0].axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5)

axs[1, 1].set_title('f(x) = x**3')
axs[1, 1].plot(x, y3)
axs[1, 1].axhline(y = 0, color = 'black', linestyle = '--', linewidth = 0.5)
axs[1, 1].axvline(x = 0, color = 'black', linestyle = '--', linewidth = 0.5);
```
