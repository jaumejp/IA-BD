### Dimensional reduction and Visualization

- a set of techniques to transform high-dimensional data into a lower dimension while retaining as much information as possible

- divided into:
    - **projections** mainly preserve *global structure*
        examples: PCA, LaplacianMaps, EigenMaps, SamonMaps
    - **embeddings** mainly preserve *local structure*
        examples: t-SNE, UMAP

### Principal Component Analysis

- PCA is extremely useful when working with data sets that have a lot of features. Common applications such as image processing, genome research always have to deal with thousands, if not tens of thousands of columns.
- PCA understands which part of our data is more informative and can summarize it in a way that we get the big picture allthough we loss some 'controlled' ammount of information.
